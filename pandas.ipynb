# Percentiles

```python
df = pd.DataFrame(dict(A=['coke', 'sprite', 'coke', 'sprite', 'sprite', 'coke', 'coke',
                          'coke', 'sprite', 'coke', 'sprite', 'sprite', 'coke', 'coke'],
                       B=['alpha','gamma', 'alpha', 'beta', 'gamma', 'beta', 'beta', 
                          'alpha','gamma', 'alpha', 'beta', 'gamma', 'beta', 'beta'],
                       col_1=[1,2,3,4,5,6,7,1,6,2,4,7,9,3]
                       ))
df
```

```python
percentiles = [50, 80, 95] # must be between 0 and 100 inclusive
cut_points = [np.percentile(df['col_1'], i) for i in percentiles]
df['group'] = 1
for i in range(len(percentiles)):
    df['group'] = df['group'] + (df['col_1'] < cut_points[i])
df
```

# Group By with Multiple Aggregations

```python
df = pd.DataFrame(dict(A=['coke', 'sprite', 'coke', 'sprite',
                          'sprite', 'coke', 'coke'],
                       B=['alpha','gamma', 'alpha', 'beta',
                          'gamma', 'beta', 'beta'],
                       col_1=[1,2,3,4,5,6,7],
                       col_2=[1,6,2,4,7,9,3]))
df
```

```python
tbl = df.groupby(['A','B']).agg({'col_1': ['max', np.mean],
                                 'col_2': ['sum','min','count']})
tbl
```

# List All Unique Values in a Group of Columns

```python
df = pd.DataFrame(dict(A=['A','A','A','A','A','B','B','B','B'],
                       B=[1,1,1,2,2,1,1,1,2],
                       C=['CA','NY','CA','FL','FL',     
                          'WA','FL','NY','WA']))
df
```

```python
tbl = df[['A', 'B', 'C']].drop_duplicates()\
                         .groupby(['A','B'])['C']\
                         .apply(list)\
                         .reset_index()
tbl
```

```python
# list to string (separated by commas) 
tbl['C'] = tbl.apply(lambda x: ('; '.join([str(s) for s in x['C']])), axis = 1)
tbl
```

# Add Row Total and Column Total to a Numerical Dataframe

```python
df = pd.DataFrame(dict(A=[2,6,3],
                       B=[2,2,6], 
                       C=[3,2,3]))
df
```

```python
df['col_total'] = df.apply(lambda x: x.sum(), axis=1)
df
```

```python
df.loc['row_total'] = df.apply(lambda x: x.sum())
df
```

# Crosstab

```python
df = pd.DataFrame(
dict(departure=['SFO', 'SFO', 'LAX', 'LAX', 'JFK', 'SFO'],
     arrival=['ORD', 'DFW', 'DFW', 'ATL', 'ATL', 'ORD'],
     airlines=['Delta','JetBlue','Delta','AA','SouthWest',  
               'Delta']),
columns=['airlines', 'departure','arrival'])
df
```

```python
pd.crosstab(index=[df['departure'], df['airlines']],
            columns=[df['arrival']],
            margins=True) # adding subtotal
```
