```python
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
```

```python
def df_preview(df, top_n_rows=5):
    print('Shape: ', df.shape)
    return df.head(top_n_rows)

df_preview(df)
```

```python
def df_info(df, top_n_rows=5):
    d = {'col_name' : [],
         'Num Nulls' : [],
         'Type' : [],
         'Num Unique' : [],
         'Sample Values' : []}
    for col_name in df.columns:
        col = df[col_name]
        uniq = col.unique()
        d['col_name'].append(col_name)
        d['Num Nulls'].append(col.isna().sum())
        d['Type'].append(col.dtypes)
        d['Num Unique'].append(uniq.shape[0])
        d['Sample Values'].append(uniq[:2])
    print('Shape: ', df.shape) 
    return pd.DataFrame(d)

df_info(df)
```

```python
def rename_col(df, old_name, new_name):
    new_columns = df.columns.values
    col_ind = list(new_columns).index(old_name)
    new_columns[col_ind] = new_name
    df.columns = new_columns
    
rename_col(df, 'CRIM', 'Some_Name')
df.head()
```
